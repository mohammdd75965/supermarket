```python
date["Date"]=pd.to_datetime(date["Date"])
```

```python
date["Date"].dtype
```

```python
date['day'] = date['Date'].dt.day.astype('object')

date['month'] = date['Date'].dt.month.astype('object')

date['year'] = date['Date'].dt.year.astype('object')
date.apply(pd.Series.nunique)
```

```python
date.drop(["year"],axis=1, inplace=True)
print(date.describe())
date.drop(["Invoice ID"],axis=1,inplace=True)
date.drop(["Tax 5%"],axis=1,inplace=True)
date.drop(["gross margin percentage"],axis=1,inplace=True)
date["Time"]=pd.to_datetime(date["Time"])
date["Houer"]=date['Time'].dt.hour.astype('object')
date.drop(["Date"],axis=1,inplace=True)
date.drop(["Time"],axis=1,inplace=True)
```

```python
plt.figure(figsize = (14,6))
ax = sns.countplot(x="Payment", hue = "Branch", data = date, palette= "tab20")
ax.set_title(label = "Payment distribution in all branches", fontsize= 25)
ax.set_xlabel(xlabel = "Payment method", fontsize = 16)
ax.set_ylabel(ylabel = "Peple Count", fontsize = 16)
```
